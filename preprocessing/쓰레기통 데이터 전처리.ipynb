# 설치지점이 "버스정류장"인 데이터에서 정류장번호 알아내기
- 상세주소에서 @@@@@ 또는 @@-@@@ 형식을 가진 데이터를 정류장 번호
- 앞 뒤로 숫자가 더 있는 경우 제외
- 첫번째 글자가 0, 1, 2 가 아닌 경우 제외





```python
import re # 정규표현 처리 라이브러리


# 정규 표현식 패턴 정의 @@@@@ 또는 @@-@@@
pattern = r'(\d{5})|(\d{2}-\d{3})'


def extract_bus_stop_number(row):
    address = row['상세주소']
    if isinstance(address, str):  # 데이터가 문자열인지 확인
        match = re.search(pattern, address)  # address에서 정규표현식 pattern 과 일치하는 부분이 있다면 그 부분을 추출
        if match:
            start_index = match.start()
            if start_index != 0 and address[start_index-1].isdigit() :
              print("상세 주소에 해당 패턴 없음 : ", address)
              return None

            last_index = match.end()
            if last_index != len(address) and address[last_index].isdigit() :
              print("상세 주소에 해당 패턴 없음 : ", address)
              return None
            if match.group()[0] not in ['0', '1', '2']:
              print(row)
              print("정류소 번호가 아님 : ", match.group())
              return None

            bus_stop_number = match.group()  # 패턴에서 일치하여 추출한 문자열 반환
            # '-' 기호가 있는 경우 제거하고 '@@@@@' 형식으로 변환
            bus_stop_number = bus_stop_number.replace('-', '')
            return '{:05}'.format(int(bus_stop_number)) # 앞부분이 0으로 채워지도록 포맷팅
    return None  # 일치하는 패턴이 없거나 데이터가 문자열이 아닌 경우 None을 반환

# '정류장번호' 열에 함수를 적용하여 정류장번호 알아내기
df['정류장번호'] = df.apply(extract_bus_stop_number, axis=1) # 기존의 Column 값을 연산하여 새 Column을 추가하는 경
# axis =0 은 row, axis = 1은 column

df
```

# 버스 정류장 번호를 이용해서 위도, 경도 알아내기
- 서울시 버스정류소 위치정보 csv
- https://data.seoul.go.kr/dataList/OA-15067/S/1/datasetView.do#


```python

# 정류장번호를 기반으로 정류소 정보를 찾아서 경도와 위도를 추가합니다.
for index, row in df.iterrows():
    정류장번호 = row['정류장번호']
    # bus_stop_data.csv에서 정류장번호에 해당하는 행을 찾습니다.
    bus_stop_info = bus_stop_data[bus_stop_data['정류소번호'] == 정류장번호]
    # 정류소 정보가 있는 경우 해당 정보의 경도(X좌표)와 위도(Y좌표)를 가져와 업데이트합니다.
    if not bus_stop_info.empty:
        df.at[index, '경도'] = bus_stop_info.iloc[0]['X좌표']
        df.at[index, '위도'] = bus_stop_info.iloc[0]['Y좌표']

df

# 업데이트된 데이터를 새로운 CSV 파일로 저장
# df.to_csv('updated_trashcan_data.csv', index=False, encoding='utf-8')

# print("업데이트된 데이터가 updated_trashcan_data.csv 파일로 저장되었습니다.")
```

# 정류장 번호가 없는 버스정류장 위도, 경도 알아내기(1)
- 상세주소와 정류소명이 같은 데이터를 찾아내서 위도, 경도 추출




```python
update_count = 0  # 업데이트된 데이터의 개수를 저장할 변수

# 정류장 번호가 존재하지 않는 데이터의 상세주소를 기반으로 정류소명과 일치하는 데이터를 찾아 X좌표를 경도, Y좌표를 위도에 입력
for index, row in df.iterrows():
    정류장번호 = row['정류장번호']
    상세주소 = row['상세주소']

    # 정류장 번호가 존재하지 않는 경우
    if pd.isnull(정류장번호):
        # 상세주소를 기반으로 정류소명과 일치하는 데이터 찾기
        matching_bus_stop = bus_stop_data[bus_stop_data['정류소명'] == 상세주소]
        # 일치하는 데이터가 있을 경우에만 경도와 위도 업데이트
        if not matching_bus_stop.empty:
            df.at[index, '경도'] = matching_bus_stop.iloc[0]['X좌표']
            df.at[index, '위도'] = matching_bus_stop.iloc[0]['Y좌표']
            update_count += 1  # 업데이트된 데이터의 개수를 증가

print(f"총 {update_count}개의 데이터가 업데이트되었습니다.")
```

# 정류장 번호가 없는 버스정류장 위도, 경도 알아내기(2)
- 상세주소에서 버스정류장 앞에 존재하는 단어와 정류소명과 동일한 데이터를 찾아내 위도, 경도 추출





```python
# 업데이트된 데이터의 개수를 저장할 변수 초기화
updated_count = 0

# trashcan_data에서 설치지점이 "버스정류장"이고 위도, 경도가 존재하지 않는 데이터 처리
for index, row in df.iterrows():
    설치지점 = row['설치지점']
    위도 = row['위도']
    경도 = row['경도']
    상세주소 = row['상세주소']

    # 설치지점이 "버스정류장"이고 위도, 경도가 존재하지 않는 경우
    if 설치지점 == '버스정류장' and pd.isnull(위도) and pd.isnull(경도):
        # 상세주소에서 "버스정류장"이라는 단어가 포함되어 있는 경우 처리
        address_parts = 상세주소.split()
        if "버스정류장" in address_parts:
            # 띄어쓰기로 split하여 "버스정류장" 문자열의 인덱스 -1 앞의 인덱스 문자열 추출
            bus_stop_index = address_parts.index("버스정류장")
            bus_stop_name = address_parts[bus_stop_index-1]
            if bus_stop_name and bus_stop_name[0] == '(':
               bus_stop_name = bus_stop_name[1:]  # 첫 번째 문자 '(' 제거
            # bus_stop_data에서 정류소명과 일치하는 데이터 찾기
            matching_bus_stop = bus_stop_data[bus_stop_data['정류소명'] == bus_stop_name]
            # 일치하는 데이터가 있을 경우에만 경도와 위도 업데이트
            if not matching_bus_stop.empty:
                print(bus_stop_name)
                df.at[index, '경도'] = matching_bus_stop.iloc[0]['X좌표']
                df.at[index, '위도'] = matching_bus_stop.iloc[0]['Y좌표']
                updated_count += 1  # 업데이트된 데이터의 개수 증가

# 업데이트된 데이터의 개수 출력
print(f"총 {updated_count}개의 데이터가 업데이트되었습니다.")
```
